```python
import json

def 合并为_json(目录, 输出文件):
    txt_files = sorted(目录.glob("*.txt"))
    合并内容 = {}

    for txt_file in txt_files:
        try:
            with open(txt_file, "r", encoding="utf-8") as f:
                合并内容[txt_file.name] = f.read()
        except Exception as e:
            print(f"读取文件 {txt_file} 时出错: {e}")

    with open(输出文件, "w", encoding="utf-8") as json_file:
        json.dump(合并内容, json_file, ensure_ascii=False, indent=4)

    print(f"所有文件已合并到 JSON 文件: {输出文件}")

if __name__ == "__main__":
    输出路径 = Path("/Users/sagawithme/Downloads/tt/merged_transcripts.json")
    合并为_json(TRANS_DIR, 输出路径)
```
